### **Random Walk with Periodic Boundaries**

1. S[n] plot represenataion for single path

> Herein, taking N=5, i.e.,

if S[n]>5, put S[n]=-2N+S[n]

if S[n]<-5, put S[n]=2N+S[n]

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import random
X=[1,-1]
temp=[]
n=[]
for i in range(100):
  x=random.choice(X)
  temp.append(x)
  n.append(i)
print("Random Choice Matrix")
print(temp)
temp_arr=np.array(temp)
Sn= np.cumsum(temp_arr)

#defining boundaries for S[n]
for i in range(len(Sn)):
  if Sn[i]>5:
    Sn[i]=-(2*5)+Sn[i]
  elif Sn[i]<-5:
    Sn[i]=(2*5)+Sn[i]
  elif Sn[i]==5:
    Sn[i]=-5
  elif Sn[i]==-5:
    Sn[i]=5

print("Cumulative Sum Matrix")
print(Sn)
# plotting for single path
plt.plot(n,Sn,label='random walk Periodic boundaries')
plt.title("Line Graph Representation")
plt.xlabel("n")
plt.ylabel("S[n]")
plt.legend();
```

2. S[n] Histogram representation

```python
fig, hist = plt.subplots(figsize=(5,5))
hist.bar(n,Sn,label='random walk periodic boundaries');
plt.xlabel("n")
plt.ylabel("S[n]")
plt.title("Histogram Representation")
plt.legend();
```

4.(i) Generate histograms for S[10] for 100,1000,10000 sample paths.

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import random
X=[1,-1]

#generating w*w matrix
def walks_(w):
  n=[]
  d=[]
  for j in range(w):
    temp=[]
    for i in range(100):
      x=random.choice(X)
      temp.append(x)
    temp_arr=np.array(temp)
    Sn=np.cumsum(temp_arr)
    n.append(Sn)
#defining boundaries for S[n]
  for j in range(len(n)):
    for i in range(len(n[j])):
        if n[j][i]>5:
          n[j][i]=-(2*5)+n[j][i]
        elif n[j][i]<-5:
          n[j][i]=(2*5)+n[j][i]
        elif n[j][i]==5:
          n[j][i]=-5
        elif n[j][i]==-5:
          n[j][i]=5
  
#for extracting 10th column, i,e. S[10]
  for j in range(len(n)):
    d.append(n[j][9])

  return(d)

#calling function for different values of w
d1=walks_(100)
d2=walks_(1000)
d3=walks_(10000)

#plotting the histogram for n=100,1000,10000
plt.show()
plt.hist(d1,15,label='for n=100')
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Representation of S[10] of 100 samples")
plt.legend();

plt.show()
plt.hist(d2,20,label='for n=1000')
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Representation of S[10] of 1000 samples")
plt.legend();

plt.show()
plt.hist(d3,20,label='for n=10000')
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Representation of S[10] of 10000 samples")
plt.legend();
```
