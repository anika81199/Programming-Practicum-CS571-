## **Integer Random walks**

1. S[n] plot represenataion for single path

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import random
X=[1,-1]
temp=[]
n=[]
for i in range(100):
  x=random.choice(X)
  temp.append(x)
  n.append(i)
print("Random Choice Matrix")
print(temp)
temp_arr=np.array(temp)
print(" ")
flag= np.cumsum(temp_arr)
print("Cumulative Sum Matrix")
print(flag)
print(" ")
plt.plot(n,flag,label='random walk int')
plt.title("Line Graph Representation")
plt.xlabel("n")
plt.ylabel("S[n]")
plt.legend();
```

2. S[n] Histogram representation

```python
fig, hist = plt.subplots(figsize=(5, 2.7))
hist.bar(n,flag,label='random walk int');
plt.xlabel("n")
plt.ylabel("S[n]")
plt.title("Histogram Representation")
plt.legend();
```

3. Random walk int for 5 different samples

```python
def arr_func():
  X=[1,-1]
  temp=[]
  n=[]
  for i in range(100):
    x=random.choice(X)
    temp.append(x)
    n.append(i)
  temp_arr=np.array(temp)
  flag= np.cumsum(temp_arr)
  return(flag)

flag1= arr_func()
flag2= arr_func()
flag3= arr_func()
flag4= arr_func()
flag5= arr_func()

plt.plot(n,flag1,label='data1')
plt.plot(n,flag2,label='data2')
plt.plot(n,flag3,label='data3')
plt.plot(n,flag4,label='data4')
plt.plot(n,flag5,label='data5')

plt.title("Line Graph Represenattion")
plt.xlabel("n")
plt.ylabel("S[n]")
plt.legend();
```

4.(i) Generate histograms for S[10] for 100,1000,10000 sample paths.

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import random
X=[1,-1]

#generating w*w matrix
def walks_(w):
  n=[]
  d=[]
  for j in range(w):
    temp=[]
    for i in range(w):
      x=random.choice(X)
      temp.append(x)
    temp_arr=np.array(temp)
    flag= np.cumsum(temp_arr)
    n.append(flag)
#for extracting 10th column, i,e. S[10]
  for j in range(len(n)):
   d.append(n[j][9])
  return(d)

#calling function for different values of w
d1=walks_(100)
d2=walks_(1000)
d3=walks_(10000)

#plotting the histogram for n=100,1000,10000
plt.show()
plt.hist(d1,50)
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Represenattion of S[10] of 100 samples")
plt.show()
plt.hist(d2,50)
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Represenattion of S[10] of 1000 samples")
plt.show()
plt.hist(d3,50)
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Represenattion of S[10] of 10000 samples")

```
