## **Continuos Random Walks for Exponential Distribution**

1. S[n] plot represenataion for single path

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import random
temp=[]
n=[]
for i in range(100):
  x=np.random.exponential()
  temp.append(x)
  n.append(i)
temp_arr=np.array(temp)
flag= np.cumsum(temp_arr)
print(flag)

#plotting line graph for exponential distribution
plt.plot(n,flag,label='random walk exponential')
plt.title("Line Graph Represenattion")
plt.xlabel("n")
plt.ylabel("S[n]")
plt.legend();
```

2. Histogram of S[n] for a single sample path

```python
# Histogram Representation of exponential distribution

plt.hist(flag,100,label='random walk exponential')
plt.xlabel("n")
plt.ylabel("S[n]")
plt.title('Histogram Representation of Random walk of exp. distribution')
plt.legend();
plt.show()
```

3. S[n] with respect to n for 5 sample paths.

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import random

def exp_func():
  temp=[]
  n=[]
  for i in range(100):
    x=np.random.exponential()
    temp.append(x)
    n.append(i)
  temp_arr=np.array(temp)
  flag= np.cumsum(temp_arr)
  return(flag)

temp1= exp_func()
temp2= exp_func()
temp3= exp_func()
temp4= exp_func()
temp5= exp_func()

plt.plot(n,temp1,label='data1')
plt.plot(n,temp2,label='data2')
plt.plot(n,temp3,label='data3')
plt.plot(n,temp4,label='data4')
plt.plot(n,temp5,label='data5')

plt.title("Line Graph Represenattion of 5 different sample paths")
plt.xlabel("n")
plt.ylabel("S[n]")
plt.legend();
```

4.(i) Generate histograms for S[10] for 100,1000,10000 sample paths.

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import random

#generating w*w matrix
def walks_(w):
  n=[]
  d=[]
  for j in range(w):
    temp=[]
    for i in range(w):
      x=np.random.exponential()
      temp.append(x)
    flag= np.cumsum(temp)
    n.append(flag)

#for extracting 10th column, i,e. S[10]
  for j in range(len(n)):
   d.append(n[j][9])
  return(d)

#plotting the histogram for n=100
d1=walks_(100)
plt.show()
plt.hist(d1,50,label='for n=100')
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Represenattion of S[10] of 100 samples")
plt.legend();
```

```python
#plotting the histogram for n=1000
d2=walks_(1000)
plt.show()
plt.hist(d2,25,label='for n=1000')
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Represenattion of S[10] of 1000 samples")
plt.legend();

#plotting the histogram for n=10000
d3=walks_(10000)
plt.show()
plt.hist(d3,50,label='for n=10000')
plt.xlabel("n")
plt.ylabel("S[10]")
plt.title("Histogram Represenattion of S[10] of 10000 samples")
plt.legend();
```
